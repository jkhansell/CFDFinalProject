```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
airfoil = np.loadtxt("modified_airfoils/airfoil1.txt", delimiter=",")
```

```python
xmax = airfoil[:,0].max()
xmin = airfoil[:,0].min()
ymax = airfoil[:,1].max()
ymin = airfoil[:,1].min()
```

```python
x = (airfoil[:,0]-xmin)/(xmax-xmin)
y = (-airfoil[:,1]+ymin)/(xmax-xmin)
y = y - np.mean(y)

plt.plot(x,y)
plt.axis("equal")
plt.scatter(np.mean(x), np.mean(y))
newairfoil = np.array([x,y]).T
np.savetxt("sg6043altered.txt",newairfoil, delimiter=",")
```
